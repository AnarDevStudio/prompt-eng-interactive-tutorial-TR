## Kurulum

API anahtarınızı yüklemek ve `get_completion` yardımcı fonksiyonunu oluşturmak için aşağıdaki kurulum hücresini çalıştırın.

```python
%pip install anthropic --quiet

# utils paketinden hints modülünü içe aktarın
import os
import sys
module_path = ".."
sys.path.append(os.path.abspath(module_path))
from utils import hints

# Python'un yerleşik düzenli ifade (regular expression) kütüphanesini içe aktarın
import re
from anthropic import AnthropicBedrock

%store -r MODEL_NAME
%store -r AWS_REGION

client = AnthropicBedrock(aws_region=AWS_REGION)

def get_completion(prompt, system=''):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ],
        system=system
    )
    return message.content[0].text
```

## Ders

**Claude, açık ve doğrudan talimatlara en iyi yanıtı verir.**

Claude'u işe yeni başlamış herhangi bir insan gibi düşünün. **Claude'un**, ona kelimesi kelimesine söyledikleriniz dışında ne yapacağına dair **hiçbir bağlamı (context) yoktur**. Bir insana bir görevi ilk kez anlatırken olduğu gibi, Claude'a tam olarak ne istediğinizi ne kadar doğrudan bir şekilde açıklarsanız, Claude'un yanıtı o kadar iyi ve doğru olacaktır.

Şüpheye düştüğünüzde, **Açık Prompt Hazırlamanın Altın Kuralı**'nı uygulayın:
- Hazırladığınız prompt'u bir meslektaşınıza veya arkadaşınıza gösterin ve istediğiniz sonucu elde edip edemediklerini görmek için talimatları bizzat takip etmelerini isteyin. Eğer onların kafası karışırsa, Claude'un da kafası karışır.

### Örnekler

Şiir yazmak gibi bir görevi ele alalım. (Hece uyumsuzluklarını görmezden gelin; LLM'ler henüz hece sayma konusunda pek başarılı değildir.)

```python
# Prompt
PROMPT = "Write a haiku about robots."

# Claude'un yanıtını yazdır
print(get_completion(PROMPT))
```

Bu haiku yeterince güzel, ancak kullanıcılar Claude'un "İşte bir haiku" gibi bir giriş cümlesi (preamble) kurmadan doğrudan şiire geçmesini isteyebilir.

Bunu nasıl sağlarız? Bunu ondan **isteyerek**!

```python
# Prompt
PROMPT = "Write a haiku about robots. Skip the preamble; go straight into the poem."

# Claude'un yanıtını yazdır
print(get_completion(PROMPT))
```

İşte başka bir örnek. Claude'a tüm zamanların en iyi basketbolcusunun kim olduğunu soralım. Aşağıda, Claude'un birkaç isim listelemesine rağmen **kesin bir "en iyi" yanıtı vermediğini** görebilirsiniz.

```python
# Prompt
PROMPT = "Who is the best basketball player of all time?"

# Claude'un yanıtını yazdır
print(get_completion(PROMPT))
```

Claude'un kararını vermesini ve en iyi oyuncunun kim olduğuna karar vermesini sağlayabilir miyiz? Evet! Sadece bunu ondan isteyin!

```python
# Prompt
PROMPT = "Who is the best basketball player of all time? Yes, there are differing opinions, but if you absolutely had to pick one player, who would it be?"

# Claude'un yanıtını yazdır
print(get_completion(PROMPT))
```

### Egzersiz 2.1 - İspanyolca
Claude'un yanıtını İspanyolca vermesini sağlamak için `SYSTEM_PROMPT` kısmını düzenleyin.

```python
# Sistem promptu - degistire bileceyin tek yer
SYSTEM_PROMPT = "[bu yaziyi degistirin]"

# Prompt
PROMPT = "Merhaba Claude nasilsin"

# claude cevabini aliyoruz
response = get_completion(PROMPT, SYSTEM_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    return "hola" in text.lower()

#  claude cevabini print ediyoruz ve doğruluğunu değerlendirmek
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("bu alisdirma cozulmusdur:", grade_exercise(response))
```

### Egzersiz 2.2 - Sadece Tek Bir Oyuncu

`PROMPT` kısmını öyle düzenleyin ki Claude hiç kaçamak cevap vermesin ve **BAŞKA HİÇBİR KELİME VEYA NOKTALAMA İŞARETİ KULLANMADAN SADECE** tek bir oyuncunun adını söylesin.

```python
# Prompt - değiştirebileceğiniz tek alan burasıdır
PROMPT = "[Bu metni değiştirin]"

# Claude'un yanıtını al
response = get_completion(PROMPT)

# Egzersiz doğruluğunu değerlendiren fonksiyon
def grade_exercise(text):
    return text == "Michael Jordan"

# Claude'un yanıtını ve karşılık gelen notu yazdır
print(response)
print("\n--------------------------- DEĞERLENDİRME ---------------------------")
print("Bu egzersiz doğru şekilde çözüldü:", grade_exercise(response))
```

### Egzersiz 2.3 - Bir Hikaye Yaz

`PROMPT` kısmını öyle düzenleyin ki Claude yazabildiği en uzun yanıtı yazsın. Eğer yanıtınız **800 kelimenin üzerindeyse**, Claude'un yanıtı doğru olarak kabul edilecektir.

```python
# Prompt - değiştirebileceğiniz tek alan burasıdır
PROMPT = "[Bu metni değiştirin]"

# Claude'un yanıtını al
response = get_completion(PROMPT)

# Egzersiz doğruluğunu değerlendiren fonksiyon
def grade_exercise(text):
    trimmed = text.strip()
    words = len(trimmed.split())
    return words >= 800

# Claude'un yanıtını ve karşılık gelen notu yazdır
print(response)
print("\n--------------------------- DEĞERLENDİRME ---------------------------")
print("Bu egzersiz doğru şekilde çözüldü:", grade_exercise(response))
```
